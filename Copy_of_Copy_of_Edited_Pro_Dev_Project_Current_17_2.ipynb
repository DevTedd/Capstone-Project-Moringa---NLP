```python
#let's now create new columns with word count, character count, sentence count, average word length, average sentence length. 
df['word_count'] = df["clean_tweets"].apply(lambda x: len(str(x).split(" ")))
df['char_count'] = df["clean_tweets"].apply(lambda x: sum(len(word) for word in str(x).split(" ")))
df['sentence_count'] = df["clean_tweets"].apply(lambda x: len(str(x).split(".")))
df['avg_word_length'] = df['char_count'] / df['word_count']
df['avg_sentence_length'] = df['word_count'] / df['sentence_count']
df.head()
```

```python
#let's represent that graphically
x, y = "word_count", "label"
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.suptitle(x, fontsize=12)
for i in df[y].unique():
    sns.distplot(df[df[y]==i][x], hist=True, kde=False, 
                 bins=10, hist_kws={"alpha":0.8}, 
                 axlabel="histogram", ax=ax[0])
    sns.distplot(df[df[y]==i][x], hist=False, kde=True, 
                 kde_kws={"shade":True}, axlabel="density",   
                 ax=ax[1])
ax[0].grid(True)
ax[0].legend(df[y].unique())
ax[1].grid(True)
plt.show()
```

Most tweets are less than 20 words long but it would seem that positive tweets are generally longer.

```python
#vectorizer = TfidfVectorizer(max_features=5000)
tf_idf = TfidfVectorizer(max_features = 8000)
# TF-IDF feature matrix
df['clean_tweets'] = df['clean_tweets'].apply(lambda x: ' '.join(x))

X = tf_idf.fit_transform(df['clean_tweets'] )
X
```

```python
X = X.toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df['label'], test_size=0.2, random_state=69)
model = GaussianNB().fit(X_train, y_train)
predicted = model.predict(X_test)
report = classification_report( y_test, predicted )
print(report)
acc=accuracy_score(y_test,predicted)
```
